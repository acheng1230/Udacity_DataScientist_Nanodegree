```python
def get_games():
    today = datetime.datetime.now()
    scoreboard = Scoreboard(today.month, today.day, today.year)
    
    # Grab today's game information
    game_header = scoreboard.game_header()
    line_score = scoreboard.line_score()
    
    # Simplify columns
    today_games = game_header[['GAME_SEQUENCE', 'GAME_ID', 'GAME_STATUS_TEXT', 
                               'LIVE_PERIOD', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID']]
    today_score = line_score[['GAME_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'PTS']]
    
    # Dictionary of game values
    team_info = today_score.set_index('TEAM_ID')['TEAM_ABBREVIATION'].to_dict()
    team_pts = today_score.set_index('TEAM_ID')['PTS'].to_dict()
    
    # Map to games info
    today_games['HOME_TEAM_ABBREVIATION'] = today_games.HOME_TEAM_ID.map(team_info)
    today_games['VISITOR_TEAM_ABBREVIATION'] = today_games.VISITOR_TEAM_ID.map(team_info)
    today_games['HOME_TEAM_PTS'] = today_games.HOME_TEAM_ID.map(team_pts)
    today_games['VISITOR_TEAM_PTS'] = today_games.VISITOR_TEAM_ID.map(team_pts)
    
    return today_games
```
